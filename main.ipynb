```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import TfidfVectorizer
import api_calls
```

```python
df = pd.read_csv('movies.csv')
df.head(5)
```

```python
df.shape
```

```python
#test
pd.set_option('display.max_colwidth', None)
movie = df[df.title == 'Iron Man']
data = movie['genres'] +" "+ movie['overview'] +" "+ movie['cast'] +" "+ movie['director']
print(data)
```

```python
df.genres.head(50)
```

```python
features = ['genres', 'overview', 'cast', 'director']

for feature in features:
    df[feature] = df[feature].fillna('')
```

```python
df.isnull().sum()
```

```python
features_text = df['genres'] +" "+ df['overview'] +" "+ df['cast'] +" "+ df['director']
features_text
```

```python
vectorizer = TfidfVectorizer()

vectorized_features = vectorizer.fit_transform(features_text)
print(vectorized_features)
```

```python
movie_name = input("Enter Movie Name")

features_string = api_calls.get_movie_features(movie_name=movie_name)
vectorized_movie = vectorizer.transform([features_string])
cosine_similarities = cosine_similarity(vectorized_movie, vectorized_features)
most_similar_indices = cosine_similarities.argsort()[0][::-1]

print(f"Recommended Movies if you've watched {movie_name.title()}:")
for idx, i in enumerate(range(1, 11)):
    print(f"{idx+1}: {df.iloc[most_similar_indices[i]]['title']}")
```
